```python
df.columns = df.columns.str.strip().str.upper().str.replace("_"," ")
df
```

```python
df.columns
```

```python
df["MONTH ADDED"] = pd.to_datetime(df["DATE ADDED"],errors = "coerce").dt.month
df["YEAR ADDED"] = pd.to_datetime(df["DATE ADDED"],errors = "coerce").dt.year
del df["DATE ADDED"]
df
```

```python
df.isna().sum()
```

```python
df["SHOW ID"].value_counts()
```

```python
df["COUNTRY"] = df["COUNTRY"].fillna("Unknown")


df["MONTH ADDED"] = df["MONTH ADDED"].fillna("Not Available")


df["YEAR ADDED"] = df["YEAR ADDED"].fillna("Not Available")

df["CAST"] = df["CAST"].fillna("Unknown")

df["DIRECTOR"] = df["DIRECTOR"].fillna("Unknown")

value3 = df["RATING"].all().mean()
df["RATING"] = df["RATING"].fillna(value3)


df
```

```python
df.isna().sum()  # we kept the duration without cleaning like removing missing values
```

```python
comparison = df["TYPE"].value_counts()
comparison.plot(kind = "bar")
plt.title("Movies vs TV Shows")
plt.show()
# Netflix  is more movie-heavy than series-heavy.
```

```python
year = df["YEAR ADDED"].value_counts()
year.plot(kind = "line")
plt.title("Content Added per Year")
plt.show()
# Netflix is  accelerating content addition .
```

```python
graph = df["YEAR ADDED"].groupby(df["TYPE"]).value_counts()
graph.plot(kind = "line")
plt.title("Movies vs TV Shows Over Time")
plt.show()
# Is Netflix shifting focus?  --- no
# It is both growing in movie as well as tv-shows industry but movie industry is growing more rapidly than tv show industry.
```

```python
graph2 = df["MONTH ADDED"].value_counts()
graph2.plot(kind = "bar")
plt.xlabel("Months ")
plt.ylabel("Number of content")
plt.title("Monthly Content Additions")
plt.show()
# Is content addition evenly distributed across months?
# yes, it is kinda evely distributed.
```

```python
data = df["TYPE"].groupby(df["DURATION"]).value_counts()
data.plot(kind = "hist")
plt.title("Movie Duration Distribution")
plt.show()
# Most movies have 97 min movie length.
# Most TV shows have 1 Season.
```
