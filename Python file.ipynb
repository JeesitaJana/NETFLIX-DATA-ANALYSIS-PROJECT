```python
df.columns = df.columns.str.strip().str.upper().str.replace("_"," ")
df
```

```python
df["MONTH ADDED "] = pd.to_datetime(df["DATE ADDED"],errors = "coerce").dt.month
df["YEAR ADDED "] = pd.to_datetime(df["DATE ADDED"],errors = "coerce").dt.year
df
```

```python
df["COUNTRY"] = df["COUNTRY"].fillna("Unknown")

value1 = df["DATE ADDED"].all().mode()
df["DATE ADDED"] = df["DATE ADDED"].fillna(value1)

actors = ["Sharukh Khan","Katrina Kaif"]
df["CAST"] = df["CAST"].fillna("Unknown")

df["DIRECTOR"] = df["DIRECTOR"].fillna("Unknown")

value2 = df["RATING"].all().mean()
df["RATING"] = df["RATING"].fillna(value2)

value3 = df["DURATION"].mode()
df["DURATION"] = df["DURATION"].fillna(value3)

df
```
